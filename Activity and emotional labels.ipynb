```python
for file in filenames:
    tokens = file.split('_')
    path = '/Users/nishantacharya/Downloads/BIRAFFE2/output csv/'+file
    subject = pd.read_csv(path)
    #Level 1 assignments
    if tokens[1] == '01':
        activity = [-1 for i in range(subject.shape[0])]
        subject["activityLabel"] = activity
        
        #Labelling the activity for moving around the map
        x_prev = 0
        y_prev = 0
        x_min_prev = 0
        y_min_prev = 0
        x_max_prev = 0
        y_max_prev = 0
        dc_prev = 0
        mc_prev = 0
        hc_prev = 0
        for i in range(subject.shape[0]):
            #Checking for position change
            if (subject['x'][i] != x_prev or subject['y'][i] != y_prev) and subject['deathCount'][i] == dc_prev and subject['collectedMoney'][i] == mc_prev and subject['collectedHealth'][i] == hc_prev:
                n = 0
                x_prev = subject['x'][i]
                y_prev = subject['y'][i]
                while n < 3:#Putting the labels for all values above and below by 10 rows
                    if i + n <subject.shape[0]:
                        subject['activityLabel'][i+n] = 3
                    if i - n >= 0:
                        subject['activityLabel'][i-n] = 3
                    n+= 1

            elif (subject['xMax'][i] != x_max_prev or subject['yMax'][i] != y_max_prev) and subject['deathCount'][i] == dc_prev and subject['collectedMoney'][i] == mc_prev and subject['collectedHealth'][i] == hc_prev:
                n = 0
                x_max_prev = subject['xMax'][i]
                y_max_prev = subject['yMax'][i]
                while n < 3:#Putting the labels for all values above and below by 10 rows
                    if i + n <subject.shape[0]:
                        subject['activityLabel'][i+n] = 3
                    if i - n >= 0:
                        subject['activityLabel'][i-n] = 3
                    n+= 1

            elif (subject['xMin'][i] != x_min_prev or subject['yMin'][i] != y_min_prev) and subject['deathCount'][i] == dc_prev and subject['collectedMoney'][i] == mc_prev and subject['collectedHealth'][i] == hc_prev:
                n = 0
                x_min_prev = subject['xMin'][i]
                y_min_prev = subject['yMin'][i]
                while n < 3:#Putting the labels for all values above and below by 10 rows
                    if i + n <subject.shape[0]:
                        subject['activityLabel'][i+n] = 3
                    if i - n >= 0:
                        subject['activityLabel'][i-n] = 3
                    n+= 1

            else:
                dc_prev = subject['deathCount'][i]
                mc_prev = subject['collectedMoney'][i]
                hc_prev = subject['collectedHealth'][i]
            
            
        #Labelling the data for gathering powerups
        dc_prev = 0
        mc_prev = 0
        hc_prev = 0
        for i in range(subject.shape[0]):
            #Checking for position change
            if subject['deathCount'][i] == dc_prev and (subject['collectedMoney'][i] != mc_prev or subject['collectedHealth'][i] != hc_prev):
                n = 0
                mc_prev = subject['collectedMoney'][i]
                hc_prev = subject['collectedHealth'][i]
                while n < 3:#Putting the labels for all values above and below by 10 rows
                    if i + n <subject.shape[0]:
                        subject['activityLabel'][i+n] = 2
                    if i - n >= 0:
                        subject['activityLabel'][i-n] = 2
                    n+= 1

            else:
                dc_prev = subject['deathCount'][i]

        #Labelling the data for attacking
        hc_prev = 0
        sc_prev = 0
        for i in range(subject.shape[0]):
            #Checking for position change
            if subject['shootsCounter'][i] != sc_prev or subject['hitCounter'][i] != hc_prev:
                n = 0
                sc_prev = subject['shootsCounter'][i]
                hc_prev = subject['hitCounter'][i]
                while n < 3:#Putting the labels for all values above and below by 10 rows
                    if i + n <subject.shape[0]:
                        subject['activityLabel'][i+n] = 1
                    if i - n >= 0:
                        subject['activityLabel'][i-n] = 1
                    n+= 1

        for i in range(subject.shape[0]):
            if subject['activityLabel'][i] == -1:
                subject['activityLabel'][i] = 4
        
        #Adding emotions
        e_path = '/Users/nishantacharya/Downloads/BIRAFFE2/BIRAFFE2-photo/BIRAFFE2-photo/'+tokens[0]+'-Face.csv'
        emo_f = pd.read_csv(e_path,delimiter = ';')
        
        #Adding columns in the anger background
        subject['ANGER'] = [-1 for i in range(subject.shape[0])]
        subject['CONTEMPT'] = [-1 for i in range(subject.shape[0])]
        subject['FEAR'] = [-1 for i in range(subject.shape[0])]
        subject['DISGUST'] = [-1 for i in range(subject.shape[0])]
        subject['HAPPINESS'] = [-1 for i in range(subject.shape[0])]
        subject['NEUTRAL'] = [-1 for i in range(subject.shape[0])]
        subject['SADNESS'] = [-1 for i in range(subject.shape[0])]
        subject['SURPRISE'] = [-1 for i in range(subject.shape[0])]
        
        #Checking the start of the game values
        count = 0
        for i in range(len(emo_f['GAME-TIMESTAMP'])):
            count = i
            if not math.isnan(emo_f['GAME-TIMESTAMP'][i]):
                break
        
        #Adding emotional Values
        #Finding the closest temporal neighbours
        face_game = emo_f[count:]
        face_secs = face_game['GAME-TIMESTAMP']*1000
        face_f = face_secs.apply(np.floor)
        for i in range(subject.shape[0]):
            smpl = subject['timestamp'][i]
            min_val = float('inf')
            index = -1
            j = count
            while j < count+face_f.shape[0]-1:
                if abs(smpl - face_f[j]) < min_val:
                    min_val = abs(smpl - face_f[j])
                    index = j
                j+=1
            subject['ANGER'][i] = face_game['ANGER'][index]
            subject['FEAR'][i] = face_game['FEAR'][index]
            subject['DISGUST'][i] = face_game['DISGUST'][index]
            subject['CONTEMPT'][i] = face_game['CONTEMPT'][index]
            subject['HAPPINESS'][i] = face_game['HAPPINESS'][index]
            subject['NEUTRAL'][i] = face_game['NEUTRAL'][index]
            subject['SADNESS'][i] = face_game['SADNESS'][index]
            subject['SURPRISE'][i] = face_game['SURPRISE'][index]
        
        #Saving the file
        save_path = '/Users/nishantacharya/Downloads/BIRAFFE2/output_label_emotion/'+tokens[0]+'_label_emotion.csv'
        subject.to_csv(save_path,index = False)
```
