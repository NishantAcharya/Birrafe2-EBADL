```python
# hori,vert,death,shoot,hit,collected money,collected health,sound id, sound stamp, 
#hori min,vert min,hori maz,vert mazz
hori_val = []
vert_val = []
death_val = []
shoot_val = []
hit_val = []
mny = []
hlth = []
sndID = []
sndTime = []
hori_min = []
vert_min = []
hori_max = []
vert_max = []
timestamp = []
```
