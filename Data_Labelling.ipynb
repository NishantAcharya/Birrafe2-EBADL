```python
import pandas as pd
import numpy as np
import json
```

```python
with open('sample-SUB211-Level02_Log.json', 'r') as json_file:
    json_load = json.load(json_file)
```

```python
# hori,vert,death,shoot,hit,collected money,collected health,sound id, sound stamp, 
#hori min,vert min,hori maz,vert mazz
hori_val = []
vert_val = []
death_val = []
ptch_lvl = []
dtrn_lvl = []
sndID = []
sndTime = []
hori_min = []
vert_min = []
hori_max = []
vert_max = []
timestamp = []
```

```python
json_load
```

```python
for i in range(len(json_load)):
    hori_val.append(json_load[i]['x'])
    vert_val.append(json_load[i]['y'])
    timestamp.append(json_load[i]['timestamp'])
    ptch_lvl.append(json_load[i]['pitchLevel'])
    dtrn_lvl.append(json_load[i]['distortionLevel'])
    sndID.append(json_load[i]['idOfSound'])
    sndTime.append(json_load[i]['timestampOfSound'])
    hori_min.append(json_load[i]['xMin'])
    vert_min.append(json_load[i]['yMin'])
    hori_max.append(json_load[i]['xMax'])
    vert_max.append(json_load[i]['yMax'])
```

```python
data = {'x':hori_val,'y':vert_val,'timestamp':timestamp,'pitchLevel':ptch_lvl,'distortionLevel':dtrn_lvl,'idOfSound':sndID,'timestampOfSound':sndTime,'xMin':hori_min,'yMin':vert_min,'xMax':hori_max,'yMax':vert_max}
```

```python
smpl211 = pd.DataFrame(data = data)
```
